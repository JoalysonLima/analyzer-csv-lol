## Limpeza dos Dados

### Análise de Impacto

```python
missings = df.isnull().sum().sort_values(ascending=False)
missings_percent = ((df.isnull().sum())*100/len(df)).sort_values(ascending=False)

impacto_nulos = pd.DataFrame({
    'Nulos': missings,
    'Percentual': missings_percent
})

impacto_nulos.head(10)
```

```python
impacto_nulos.describe()
```

```python
impacto_nulos.head(int((len(impacto_nulos)*0.25)//1))
```

```python
colunas_excluir = impacto_nulos[impacto_nulos['Percentual'] >= 80].index
colunas_excluir
```

```python
df_limpo = df.drop(colunas_excluir, axis=1)
```

```python
df_limpo.head()
```

```python
missings = df_limpo.isnull().sum().sort_values(ascending=False)
missings_percent = ((df_limpo.isnull().sum())*100/len(df)).sort_values(ascending=False)

impacto_nulos = pd.DataFrame({
    'Nulos': missings,
    'Percentual': missings_percent
})

impacto_nulos.head()
```
