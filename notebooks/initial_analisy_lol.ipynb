### Análise de Impacto

```python
missings = df.isnull().sum().sort_values(ascending=False)
missings_percent = ((df.isnull().sum())*100/len(df)).sort_values(ascending=False)

impacto_nulos = pd.DataFrame({
    'Nulos': missings,
    'Percentual': missings_percent
})

impacto_nulos = impacto_nulos[impacto_nulos['Nulos'] > 0]
```

```python
impacto_nulos.describe()
```

```python
impacto_nulos.head(int((len(impacto_nulos)*0.25)//1))
```

```python
colunas_excluir = impacto_nulos[impacto_nulos['Percentual'] >= 80].index
colunas_excluir
```

**Decidi excluir todas as colunas que possuiam mais de 80% de valores nulos.**

```python
df_limpo = df.drop(colunas_excluir, axis=1)
```

```python
df_limpo.head()
```

```python
object_cols = df_limpo.select_dtypes(include=['object']).columns

for col in object_cols:
    df_limpo[col] = df_limpo[col].fillna('Unknowm')
```

**As colunas de tipo `object` não apresentam impacto direto na variável dependente. Logo, elas foram excluídas.**

```python
missings = df_limpo.isnull().sum().sort_values(ascending=False)
missings_percent = ((df_limpo.isnull().sum())*100/len(df)).sort_values(ascending=False)

impacto_nulos = pd.DataFrame({
    'Nulos': missings,
    'Percentual': missings_percent
})

impacto_nulos = impacto_nulos[impacto_nulos['Nulos'] > 0]
```

```python
colunas_analise = impacto_nulos.index
colunas_analise
```

```python
df_limpo[colunas_analise].info()
```

**Todas as colunas restantes que ainda possuem valores nulos são do tipo `float64` e apresentam impacto direto ou indireto na variável dependente e precisam de uma análise mais aprofundada.**
